```python
# Week 1: Operational Efficiency Analyzer
### Target Audience: Junior Analysts / Team Leads

# This notebook automates the tracking of daily team performance across our corporate support office.
# Instead of calculating performance metrics manually, this script introduces automated tracking logic to quickly flag teams that are hitting targets, falling slightly behind, or requiring immediate assistance.

# **Key Components:**
# 1. `calculate_efficiency`: A reusable logic block that computes performance percentages.
# 2. `get_operational_status`: A classifier that applies business rules to categorize performance.
# 3. Data Modeling: A structured data setup representing our active customer support departments.
# 4. Execution Loop: A process that automatically reviews each team and prints an executive summary report.
```

```python
def calculate_efficiency(actual_output, target_output):
    """
    Calculates the efficiency percentage based on actual output vs target output.
    Prevents division by zero if target_output is set to 0.
    """
    if target_output == 0:
        return 0.0
    return (actual_output / target_output) * 100

def get_operational_status(efficiency):
    """
    Classifies operational status based on efficiency thresholds:
    - Normal: >= 90%
    - Warning: 70% to 89%
    - Critical: < 70%
    """
    if efficiency >= 90:
        return "Normal"
    elif 70 <= efficiency < 90:
        return "Warning"
    else:
        return "Critical"
```

```python
# Data Modeling: Structured dictionary representing 3 distinct customer support teams
support_teams = {
    "team_1": {
        "name": "Tier 1 Support",
        "actual_output": 95,
        "target_output": 100
    },
    "team_2": {
        "name": "Billing Inquiries",
        "actual_output": 80,
        "target_output": 100
    },
    "team_3": {
        "name": "Technical Escalations",
        "actual_output": 55,
        "target_output": 100
    }
}

# Execution: Loop through each team, process their data, and print a formatted report
print("=" * 50)
print("       DAILY OPERATIONAL EFFICIENCY REPORT       ")
print("=" * 50)

for team_key, team_data in support_teams.items():
    # Calculate efficiency
    efficiency = calculate_efficiency(team_data["actual_output"], team_data["target_output"])

    # Determine health status
    status = get_operational_status(efficiency)

    # Print clean formatted output for stakeholders
    print(f"Department: {team_data['name']}")
    print(f"  - Actual Tickets Resolved: {team_data['actual_output']}")
    print(f"  - Target Daily Goal:     {team_data['target_output']}")
    print(f"  - Efficiency Rating:     {efficiency:.1f}%")
    print(f"  - Operational Status:    [{status}]")
    print("-" * 50)
```
